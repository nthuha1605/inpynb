#### drinks.csv là tập tin cung cấp dữ liệu về tình hình tiêu thụ rượu bia ở các quốc gia theo từng châu lục

country: tên quốc gia tiêu thụ<br/>
beer_servings: số lượng bia tiêu thụ<br/>
spirit_servings: số lượng rượu mạnh (>= 40<sup>0</sup>)<br/>
wine_servings: số lượng rượu vang<br/>
total_litres_of_pure_alcohol: tổng số lít rượu nguyên chất<br/>
continent: châu lục

#### Thực hiện các yêu cầu sau, và đối chiếu với kết quả cho trước:

```python
import pandas as pd
```

```python


# Câu 1: Đọc dữ liệu từ tập tin drinks.csv với index_col là cột đầu tiên của dữ liệu, và lưu vào biến drink.
drink = pd.read_csv(r'D:\TH4\drinks.csv', index_col=0)
print(drink)

# Cho biết kiểu dữ liệu (type) của drink
print(type(drink))

# Cho biết ra kích thước (shape) của drink
print(drink.shape)

# Hiển thị tên các cột (columns) của drink
print(drink.columns)

# Xem 5 dòng dữ liệu đầu tiên (head) của drink
print(drink.head())
```

```python
# Câu 2: Xác định số lượng giao dịch theo từng châu lục (continent) bằng Groupby
drink.groupby('continent').size()
```

```python
# Câu 3:
# - Xác định bao nhiêu thuộc tính có giá trị NULL và số lượng giá trị NULL tương ứng với từng thuộc tính
gtr_null = drink.isnull().sum()
print(gtr_null)

# - Nếu có giá trị NULL liệt kê chi tiết từng giao dịch của mỗi thuộc tính
for column in drink.columns:
    null_indices = drink[drink[column].isnull()].index
    if len(null_indices) > 0:
        print(f"\nThuộc tính '{column}':")
        for index in null_indices:
            print(f"- Giao dịch tại {index}: {drink.loc[index]}")
```

```python
# Câu 4: Cho biết số lượng bia tiêu thụ trung bình ở mỗi châu lục
tb_bia_tieuthu = drink.groupby('continent')['beer_servings'].mean()
print(tb_bia_tieuthu)
```

```python
# Câu 5: Lập danh sách số lượng bia tiêu thụ giảm dần theo châu lục
bia_giam = drink.groupby('continent')['beer_servings'].mean().sort_values(ascending=False)
print(bia_giam)
```

```python
# Câu 6: Cho biết thông tin thống kê tổng quát về số lượng rượu vang được tiêu thụ ở mỗi châu lục 
ruou_vang = drink.groupby('continent')['wine_servings'].describe()
print(ruou_vang)
```

```python
# Câu 7: Cho biết thông tin thống kê tổng quát về số lượng bia và rượu mạnh. Chú ý bố trí trong cùng một bảng dữ liệu.
bia_ruoumanh = drink[['beer_servings', 'spirit_servings']].describe()
print(bia_ruoumanh)
```

```python
# Câu 8: Cho biết giá trị trung vị (median) của các loại bia và rượu tiêu thụ ở mỗi châu lục
giatritrungvi = drink.groupby('continent')[['beer_servings', 'spirit_servings']].median()
print(giatritrungvi)
```

```python
# Câu 9: Cho biết giá trị trung bình,trung vị,lớn nhất về số lượng bia tiêu thụ ở mỗi châu lục
beer = drink.groupby('continent')['beer_servings'].agg(['mean', 'median', 'max'])
print(beer)
```

```python
# Câu 10: Sắp xếp dữ liệu tăng dần theo số lượng bia tiêu thụ lớn hơn 300
tangdan = drink[drink['beer_servings'] > 300].sort_values(by='beer_servings', ascending=True)
print(tangdan)
```

```python
# Câu 11: Liệt kê danh sách quốc gia theo châu lục
country = drink.groupby('continent')['country'].apply(list)
print(country)
```

```python
# Câu 12: Cho biết 5 quốc gia có lượng tiêu thụ bia lớn nhất
top5 = drink.nlargest(5, 'beer_servings')
print(top5)
```

## Assignment 2- Pivot Table

####  - restaurant.csv là tập tin cung cấp dữ liệu về tình hình giao dịch tại một nhà hàng


```python
restaurant = pd.read_csv(r'D:\TH4\restaurant.csv', index_col=0)
print(restaurant)
# Câu 1: Tính trung bình theo total bill của nam và nữ trong các ngày trong tuần
# Đặt cột 'total_bill' thành một cột dữ liệu
restaurant.reset_index(inplace=True)
nam_nu = restaurant.groupby(['day', 'sex'])['total_bill'].mean()
print(nam_nu)
# Câu 2: Tính tổng tiền tip và total bill của nam và nữ trong các ngày trong tuần
tongtien = restaurant.groupby(['day', 'sex'])[['tip', 'total_bill']].sum()
print(tongtien)
# Câu 3: Xây dựng bảng tổng hợp thống kê liên quan tính tiền tip và total bill cho ngày (day) và giới tính (sex), biết rằng tiền
# tip thì tính theo giá trị trung bình còn total bill thì tính tổng.
thongke = restaurant.groupby(['day', 'sex']).agg({'tip':'mean', 'total_bill':'sum'})
thongke = ['average_tip', 'total_bill_sum']
print(thongke)
# Câu 4: Thêm một cột xếp loại tiền tip ở dataframe, và cột này được nhận 'high tip' nếu tiền tip của họ lớn hơn 3 và
# ngược lại thì nhận 'low tip'.
def tip_category(tip_amount):
    if tip_amount > 3:
        return 'high tip'
    else:
        return 'low tip'
    
restaurant['tip_category'] = restaurant['tip'].apply(tip_category)
print(restaurant)
# Câu 5: Thêm một cột xếp loại khách hàng và chỉ xét cho nữ, giá trị cột này được tính như sau: nếu total bill
# nhỏ hơn 20 thì nhận 'poor girl' còn ngược lại thì nhận 'rich girl'
def customer_category(row):
    if row['sex'] == 'Female':
        if row['total_bill'] < 20:
            return 'poor girl'
        else:
            return 'rich girl'
    else:
        return None
restaurant['customer_category'] = restaurant.apply(customer_category, axis=1)
print(restaurant)

# Câu 6: Xác định số lượng giao dịch vào thứ 7
gd_t7 = restaurant[restaurant['day'] == 'Sat'].shape[0]
print(gd_t7)
```
