```python
# Câu 1: Tạo và thao tác ở mảng 2D
# a) Tạo một ma trận 2 chiều có kích thước 4x4 với các giá trị ngẫu nhiên từ 1 đến 16.
import numpy as np
arr = np.random.randint(1, 17, (4, 4))
print(arr)
```

\begin{equation}
    2x+y-z   =8\\
    4x-6y+2z =-3\\
    2x+8y-3z =3

## Assignment 2 - Loại dữ liệu tự định nghĩa

Tạo một đối tượng có tên là sản phẩm với các thuộc tính như sau: mã sản phẩm (masp), tên sản phẩm (tensp), và giá sản phẩm (price), số lượng (soluong). Dữ liệu cụ thể như sau:


(1, 'Máy tính', 1200.50,10),
(2, 'Điện thoại Iphone 11', 699.99,20),
(3, 'USB', 89.95,20),
(4, 'Loa ', 499.50,30),
(5, 'Màn hình', 799.00,50)


Yêu cầu:

a) Tìm sản phẩm có giá cao nhất

b) Tính giá trung bình của tất cả sản phẩm

c) Tính doanh số bán, biết rằng doanh số bán bằng số lượng x đơn giá

d) Hiện thị thông tin sản phẩm gồm có tên sản phẩm và giá sản phẩm

```python
# a) Tìm sản phẩm có giá cao nhất
import numpy as np

# Câu 1
# Dữ liệu sản phẩm
product_data = [
    (1, 'Máy tính', 1200.50, 10),
    (2, 'Điện thoại Iphone 11', 699.99, 20),
    (3, 'USB', 89.95, 20),
    (4, 'Loa', 499.50, 30),
    (5, 'Màn hình', 799.00, 50)
]

# a) Tìm sản phẩm có giá cao nhất
products = np.array(product_data, dtype=[('masp', int), ('tensp', 'U30'), ('price', float), ('soluong', int)])

max_price_product = products[np.argmax(products['price'])]
print("Sản phẩm có giá cao nhất:")
print("Mã sản phẩm:", max_price_product['masp'])
print("Tên sản phẩm:", max_price_product['tensp'])
print("Giá sản phẩm:", max_price_product['price'])
print("Số lượng:", max_price_product['soluong'])
print()
```

```python
# b) Tính giá trung bình của tất cả sản phẩm
average_price = np.mean(products['price'])
print("Giá trung bình của tất cả sản phẩm:", average_price)
print()
```

```python
#c) Tính doanh số bán, biết rằng doanh số bán bằng số lượng x đơn giá
total_revenue = np.sum(products['price'] * products['soluong'])
print("Tổng doanh số bán:", total_revenue)
print()
```

```python
# d) Hiện thị thông tin sản phẩm gồm có tên sản phẩm và giá sản phẩm
# Hàm hiển thị thông tin chi tiết về mỗi sản phẩm
print("Thông tin sản phẩm (tên và giá):")
for product in products:
    print("Tên sản phẩm:", product['tensp'])
    print("Giá sản phẩm:", product['price'])
    print()
```
